```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df  = pd.read_csv("Expanded_data_with_more_features.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#check null values
df.isnull().sum()
```

```python
#drop null values
df1 = df.drop("Unnamed: 0", axis = 1)
df1.head()
```

```python
df["WklyStudyHours"].unique()
```

```python
 #gender distribution 
plt.figure(figsize= (4,4))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()

#from the below chart we have analysed that the no of female is more than no of male in data
```

```python
df1.head()
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb
```

```python
plt.figure(figsize = (5,4))
sns.heatmap(gb, annot = True)
plt.title("Impact in Students Scores by Parents Education Status")
plt.show()

#from the below Heatmap we can conclude that the education of the parents impact 
#the score of the students
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb1
```

```python
plt.figure(figsize = (5,4))
sns.heatmap(gb1, annot = True)
plt.title("Impact in Students Scores by Parents Marital Status")
plt.show()

#from the below chart we can see that the parents marital status won't make much impact
#to the students scores
```

```python
#DISTRIBUTION OF ETHNIC GROUPS

groupA = df.loc[(df1["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df1["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df1["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df1["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df1["EthnicGroup"] == "group E")].count()

#groupA["EthnicGroup"]
#groupB["EthnicGroup"]

l = ["A","B","C","D","E"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],
         groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
